# 4.TRAINING AND TESTING THE DATA : This is the next phase of co-relation in this we will train and test the model to do the predictions. Here we will split the data into 2 parts 1.train and 2.test. Generally we will take 80% of data to train and 20% of data to test. If we give 100 data to train it will predict the 100% accuracy but, it is not good and it makes the machine over fitting.

```python
x=data.drop(['species'],axis=1) #dropping the target column from x
```

```python
y= data['species'] #initilizing the target to  y
```

# 5. SELECTING THE  BEST ALGORITHM : This is also necessary phase to select the best algorithm to do the prediction. Based on the input and output we have to choose the algorithm. Here we choose the logistic regression because of  binary classification of output

```python
from sklearn.linear_model import LogisticRegression #importing the LogisticRegression package.
```

```python
algo = LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3,random_state=5) #training,testing and splitting the dataaset.
```

```python
algo.fit(xtrain,ytrain)
```

```python
ypred = algo.predict(xtest) #predicting the xtest based on ytest
ypred = pd.DataFrame(ypred) #converting array format to dataframe format
ypred.head()
```

# 6. CHECKING THE PEDICTIONS : In this we will check how many predictions were correct and wrong. We make use of CONFUSION_MATRIX  to do work for us.

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(ytest,ypred)
```

in this we can say 15 were correct predicions from 1st column, 13 were correct predicions from 1st column, 14 were correct predicions from 1st column.

# 7. CHECKING THE ACCURACY SCORE : This is the final phase, in this we will check the accuracy score of the ypred(features) which predicted using ytest. In this we make use of ACCURACY_SCORE package to do work for us.

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(ytest,ypred)
```
